#   1. What are Type I and Type II errors in hypothesis testing, and how do they impact decision-making?
 -  Type I and Type II errors are false conclusions in hypothesis testing. A Type I error (false positive, \(\alpha \)) occurs when a true null hypothesis (\(H_{0}\)) is wrongly rejected. A Type II error (false negative, \(\beta \)) occurs when a false \(H_{0}\) is not rejected. They impact decision-making by leading to either unnecessary actions (Type I) or missed opportunities/undetected problems (Type II), often requiring atrade-off to minimize risk.

Type I Error (False Positive - "Alpha Error")
       -  Definition: Rejecting the null hypothesis (\(H_{0}\)) when it is actually true.
 -  Significance (\(\alpha \)): The probability of committing this error is typically set to \(0.05\) or \(0.01\).
 -  Example: A test finds a new website design is better, but it actually isn't, leading to a loss in revenue.


Type II Error (False Negative - "Beta Error")
  -  Definition: Failing to reject (retaining) the null hypothesis (\(H_{0}\)) when it is actually false.
  -  Significance (\(\beta \)): The probability of failing to detect a true effect
   -  Example: A diagnostic test fails to detect a disease in a patient who actually has it.
Both errors can be reduced by increasing the sample size, which provides more data and better precision.

 p-value in hypothesis testing is used to decide whether to reject a null hypothesis.

Definition of the P-value - The P-value (probability value) quantifies the strength of evidence against a null hypothesis (\(\mathbf{H}_{\mathbf{0}}\)). It represents the probability of obtaining test results at least as extreme as the results actually observed, assuming that the \(\mathbf{H}_{\mathbf{0}}\) is correct and random chance is the sole factor at play.

Interpretation in the Context of the Null Hypothesis -The P-value is a critical metric for decision-making in hypothesis testing. The interpretation hinges on a comparison with a predetermined significance level (\(\alpha \)), typically set at \(0.05\).

  Compare P-value to \(\alpha \)

  Small P-value (\(\mathbf{P\le \alpha }\)): The result is statistically significant. This provides strong evidence against the null hypothesis, leading to the decision to reject the null hypothesis. The observed effect is unlikely to be due to chance.

  Large P-value (\(\mathbf{P>\alpha }\)): The result is not statistically significant. There is insufficient evidence to reject the null hypothesis, leading to the decision to fail to reject the null hypothesis. The observed results are reasonably likely to have occurred by random chance under the assumption that \(\mathbf{H}_{\mathbf{0}}\) is true


# 3.Explain the difference between a Z-test and a T-test, including when to use each.
  -  Z-tests and T-tests both determine if sample means differ significantly from a population mean or each other. Use a Z-test when the population standard deviation is known and sample sizes are large (\(n>30\)), relying on the normal distribution. Use a T-test when the population standard deviation is unknown, regardless of sample size (though primarily for \(n<30\)).
    Key Differences Between Z-Test and T-Test
  -  Population Standard Deviation (\(\sigma \)): A Z-test requires knowing the population standard deviation. A T-test is used when the population standard deviation is unknown, using the sample standard deviation (\(s\)) instead.
  -  Sample Size (\(n\)): Z-tests are appropriate for large samples (typically \(n\ge 30\)). T-tests are primarily used for small sample sizes (\(n<30\)), although they can be used for larger samples if the population standard deviation is unknown.
  - Distribution: Z-tests follow the standard normal distribution (\(Z\)-distribution). T-tests follow the Student’s T-distribution.
  -  Precision: Z-tests are generally more precise when data allows, while T-tests are more conservative to account for added uncertainty from estimating population standard deviation.

  Examples

 Use Z-test: You want to know if the average exam score of 100 students (sample size) is different from the national average, assuming the national standard deviation is known.
 Use T-test: You want to compare the average weight of 15 apples (small sample) from two different trees without knowing the true standard deviation of all apple weights.

# 5. Describe the purpose and assumptions of an ANOVA test. How does it extend hypothesis testing to more than two groups?
 -  The purpose of an ANOVA (Analysis of Variance) test is to determine if there are statistically significant differences between the means of three or more independent groups by comparing the variance between groups to the variance within groups. It acts as a global test to reject the null hypothesis that all group means are equal.
 Key Assumptions of ANOVA
 -  Independence of Observations: Data points are independent, with no relationship between subjects in different groups or within groups.
  - Normality: The dependent variable is approximately normally distributed within each group.
 -  Homogeneity of Variances (Homoscedasticity): The variances of the populations are equal across all groups, often checked via Levene's test.
 - Continuous Data: The dependent variable is measured at the interval or ratio level, according to Laerd Statistics.

 # 6. Write a Python program to perform a one-sample Z-test and interpret the result for a given dataset.
  -  A one-sample Z-test can be performed in Python using the ztest function from the statsmodels.stats.weightstats module or by manually calculating the Z-statistic and p-value using scipy.stats. The result is interpreted by comparing the p-value to a pre-determined significance level (), typically 0.05.
  Python Program for One-Sample Z-Test
  This program uses the statsmodels library to perform the test and a custom function to interpret the results.
  import numpy as np
from statsmodels.stats.weightstats import ztest

# Sample data: 100 smartphones, avg 11.8h life, population SD 0.5h
sample_data = np.random.normal(11.8, 0.5, 100)
pop_mean = 12.0 # Hypothesized mean
alpha = 0.05

# Perform Z-test
z_statistic, p_value = ztest(sample_data, value=pop_mean)

print(f"Z-Statistic: {z_statistic:.4f}")
print(f"P-Value: {p_value:.4f}")

# Interpretation
if p_value < alpha:
    print("Result: Reject null hypothesis - Significant difference")
else:
    print("Result: Fail to reject null hypothesis - No significant difference")

# 8. Generate multiple samples from a non-normal distribution and implement the Central Limit Theorem using Python.
  -  To demonstrate the Central Limit Theorem (CLT) using a non-normal distribution, you can use Python libraries like numpy to generate samples from a uniform or exponential distribution and matplotlib to visualize the results. The following steps outline the process and provide sample code using the uniform distribution.
  Python Implementation of the Central Limit Theorem
  The code below performs the following:
 -  Generates a large population from a non-normal uniform distribution.
 -  Repeatedly samples from this population, calculating the mean of each sample.
 -  Plots the distribution of these sample means, which will approximate a normal distribution as the sample size increases, illustrating the CLT.
 import numpy as np
import matplotlib.pyplot as plt

# --- Step 1: Define the non-normal population (Uniform Distribution) ---
# The population is non-normal and uniformly distributed between 0 and 100
population_size = 100000
population = np.random.uniform(low=0, high=100, size=population_size)
population_mean = np.mean(population)
population_std = np.std(population)
print(f"Population Mean: {population_mean:.2f}")
print(f"Population Standard Deviation: {population_std:.2f}")

# --- Step 2: Set parameters for sampling ---
sample_sizes = [2, 10, 30, 100]  # Different sample sizes to observe the effect
num_samples = 10000              # Number of samples to draw for each size
all_sample_means = {size: [] for size in sample_sizes}

# Generate sample means
for size in sample_sizes:
    for _ in range(num_samples):
        # Draw random samples with replacement
        sample = np.random.choice(population, size=size, replace=True)
        sample_mean = np.mean(sample)
        all_sample_means[size].append(sample_mean)

# --- Step 3: Plot the distributions ---
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle('Distribution of Sample Means (Central Limit Theorem Demonstration)', fontsize=16)

for ax, size in zip(axes.flatten(), sample_sizes):
    means = all_sample_means[size]
    ax.hist(means, bins=30, density=True, alpha=0.75, color='blue', edgecolor='black')
    ax.set_title(f"Sample Size = {size} (N={num_samples} samples)")
    ax.set_xlabel("Sample Mean")
    ax.set_ylabel("Density")
    ax.axvline(population_mean, color='red', linestyle='dashed', linewidth=2, label='Population Mean')
    ax.legend()
    ax.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])

# 9. Write a Python function to calculate and visualize the confidence interval for a sample mean.
-  import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt

def calculate_and_visualize_ci_mean(data, confidence_level=0.95):
    """
    Calculates the confidence interval for a sample mean and visualizes it.

    Args:
        data (list or np.array): The sample data.
        confidence_level (float): The desired confidence level (e.g., 0.95 for 95%).

    Returns:
        tuple: (sample_mean, ci_lower, ci_upper)
    """
    # Convert data to numpy array for easier manipulation
    data = np.array(data)
    n = len(data)
    sample_mean = np.mean(data)
    
    # Calculate the standard error of the mean (sem)
    # scipy.stats.sem uses a default ddof=1 which is appropriate for sample std dev
    standard_error = st.sem(data)
    degrees_freedom = n - 1

    # Calculate the confidence interval using the t-distribution
    # st.t.interval returns the lower and upper bounds
    ci_lower, ci_upper = st.t.interval(
        alpha=confidence_level,
        df=degrees_freedom,
        loc=sample_mean,
        scale=standard_error
    )
    
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"{confidence_level*100}% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
    
    # Visualization
    plt.figure(figsize=(8, 6))
    plt.hist(data, bins=10, color='skyblue', alpha=0.7, label='Sample Data Distribution')
    plt.axvline(sample_mean, color='green', linestyle='solid', linewidth=2, label=f'Sample Mean ({sample_mean:.2f})')
    plt.axvline(ci_lower, color='red', linestyle='dashed', linewidth=2, label=f'CI Bounds ({ci_lower:.2f}, {ci_upper:.2f})')
    plt.axvline(ci_upper, color='red', linestyle='dashed', linewidth=2)
    plt.xlabel('Data Values')
    plt.ylabel('Frequency')
    plt.title(f'Confidence Interval for Sample Mean (alpha={confidence_level*100}%)')
    plt.legend()
    plt.show()

    return sample_mean, ci_lower, ci_upper

# --- Example Usage ---
# Sample data (e.g., exam scores)
exam_scores = [75, 80, 85, 90, 92, 78, 88, 95, 79, 82, 81, 86, 89, 91, 84]

# Calculate and visualize the 95% confidence interval
calculate_and_visualize_ci_mean(exam_scores, confidence_level=0.95)


# 10. Perform a Chi-square goodness-of-fit test using Python to compare observed and expected distributions, and explain the outcome.
 -  A Chi-square goodness-of-fit test can be performed in Python using the scipy.stats.chisquare function. The test helps determine if observed frequencies differ significantly from expected frequencies based on a hypothesized distribution.
Python Implementation
This example tests if a six-sided die is fair. We roll it 60 times and record the observed frequencies. The expected frequency for a fair die is 10 for each face.
import numpy as np
from scipy.stats import chisquare

# Observed frequencies from the experiment (e.g., rolling a die 60 times)
observed_frequencies = np.array([9, 11, 16, 10, 9, 5])

# Expected frequencies if the die is fair (uniform distribution)
expected_frequencies = np.array([10, 10, 10, 10, 10, 10])

# Perform the Chi-square goodness-of-fit test
chi2_statistic, p_value = chisquare(f_obs=observed_frequencies, f_exp=expected_frequencies)

print(f"Chi-Square Statistic: {chi2_statistic}")
print(f"P-value: {p_value}")
